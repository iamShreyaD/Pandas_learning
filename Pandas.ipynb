```python
df1 = pd.read_csv("Data.csv")
```

```python
print(df1)
```

```python
#Selection by column
print(df1["Name"])
```

```python
pokemon = input("Enter a Pokemon name: ")
try:
    print(df1.loc[pokemon])
except KeyError:
    print(f"{pokemon} not found")
```

```python
print(df1.loc["Charizard"])
```

```python
#filtering is keeping the rows that match the condition
tall_pokemon = df1[df1["Height"] >= 2.0]
print(tall_pokemon)
```

```python
heavy_pokemon = df1[df1["Weight"] >= 200]
print(heavy_pokemon)
```

```python
legendary_pokemon = df1[df1["Legendary"] == True]
print(legendary_pokemon)
```

```python
water_pokemon = df1[(df1["Type1"] == "Water") | (df1["Type2"] == "Water")]
print(water_pokemon)
```

```python
hard_pokemon = df1[(df1["Type1"] == "Rock") | (df1["Type2"] == "Ground")]
print(hard_pokemon)
```

```python
#Data cleaning is the process of fixing/removing incomplete, incorrect, or irrelevant data
#Drop irrelevant columns
df1 = df1.drop(columns=["No"])
print(df1)
```

```python
#Handle missing data 
#dropna = drop not available

#Method 1
#df1 = df1.dropna(subset = ["Type2"])

#Method 2
df1 = df1.fillna({"Type2": "None"})
print(df1.to_string())
```

```python
#Fix inconsistent values,

df1["Type1"] = df1["Type1"].replace({"Grass":"GRASS",
                                    "Fire":"FIRE",
                                    "Water":"WATER"})
print(df1)
```

```python
#Standardize text
df1["Name"] = df1["Name"].str.lower()
print(df1)
```

```python
#Fix data types
df1["Legendary"] = df1["Legendary"].astype(bool)
print(df1)
```

```python
#Remove duplicate values
df1 = df1.drop_duplicates()
print(df1)
```
